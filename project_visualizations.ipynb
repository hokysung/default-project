```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vanilla = np.load("../vanilla_easy_res.npy")
share = np.load("../cycle_res.npy")
```

```python
baseline_fixed = 0.7440000176429749
baseline_trainable = 0.9449999928474426
```

```python
epochs = np.arange(0, 40, 2)
plt.plot(epochs, vanilla, label='Vanilla GAN')
plt.plot(epochs, share, label='ShareGAN')
plt.xlabel("GAN Training Epochs", fontsize=15)
plt.ylabel("Classification Accuracy", fontsize=15)
ax = plt.gca()
plt.hlines(baseline_fixed, 0, 40, label='Baseline (fixed random MLP)', color='red')
plt.hlines(baseline_trainable, 0 , 40, label='Baseline (trainable MLP)', color='black')
plt.legend(fontsize=15, bbox_to_anchor=(1.85,0.75))
plt.show()
```
